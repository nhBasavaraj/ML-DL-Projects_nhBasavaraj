```python

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

y_close = y['Close']
y_adj_close = y['Adj Close']

X = X.drop('Date', axis=1)

cv=ShuffleSplit(n_splits=10, test_size=0.2)
score_close=cross_val_score(LinearRegression(),X,y_close, cv=cv)
score_adj_close=cross_val_score(LinearRegression(), X, y_adj_close, cv=cv)
```

```python
print("Mean Accuracy (Close):", score_close.mean())
print("Mean Accuracy (Adj Close):", score_adj_close.mean())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

# Create separate models for each target variable
models = {
    'Close': [
        ('Linear Regression', LinearRegression(), {}),
        ('Decision Tree', DecisionTreeRegressor(), {'max_depth': [None, 10, 20]}),
        ('Random Forest', RandomForestRegressor(), {'n_estimators': [50, 100, 200]}),
        ('SVM', SVR(), {'C': [0.1, 1.0, 10.0]}),
    ],
    'Adj Close': [
        ('Linear Regression', LinearRegression(), {}),
        ('Decision Tree', DecisionTreeRegressor(), {'max_depth': [None, 10, 20]}),
        ('Random Forest', RandomForestRegressor(), {'n_estimators': [50, 100, 200]}),
        ('SVM', SVR(), {'C': [0.1, 1.0, 10.0]}),
    ]
}

# Perform GridSearchCV for each target variable and model
best_scores = {}
for target_var, models_list in models.items():
    for name, model, params in models_list:
        grid_search = GridSearchCV(model, param_grid=params, cv=5, n_jobs=-1)
        grid_search.fit(X_train, y_train[target_var])
        best_scores[(target_var, name)] = grid_search.best_score_

# Print the best scores for each target variable and model
for (target_var, name), score in best_scores.items():
    print(f"{target_var} - {name} - Best Score: {score:.4f}")
```

```python
import joblib
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

# Create separate models for each target variable
models = {
    'Close': [
        ('Linear Regression', LinearRegression(), {}),
        ('Decision Tree', DecisionTreeRegressor(max_depth=10), {}),
        ('Random Forest', RandomForestRegressor(n_estimators=100), {}),
        # SVM does not perform well for this problem, so we'll skip saving it
    ],
    'Adj Close': [
        ('Linear Regression', LinearRegression(), {}),
        ('Decision Tree', DecisionTreeRegressor(max_depth=10), {}),
        ('Random Forest', RandomForestRegressor(n_estimators=100), {}),
        # SVM does not perform well for this problem, so we'll skip saving it
    ]
}

# Train and save the best models
for target_var, models_list in models.items():
    for name, model, params in models_list:
        model.fit(X_train, y_train[target_var])
        joblib.dump(model, f"{target_var} - {name}.pkl")
```
